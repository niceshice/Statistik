```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

ger_childrens_book_authors = pd.read_csv(r"D:/cravi/Documents/Uni/DH/Statistik/Data/KinderautorenD20Jhdt.tsv", sep="\t")
ger_childrens_book_authors["Lebensspanne"] = ger_childrens_book_authors["Todesjahr"] - ger_childrens_book_authors["Geburtsjahr"]
```

#### Barplot Lebensspanne
..der Autoren aus dem eigenen Datensatz

```python
ger_childrens_book_authors.sort_values(by = "Lebensspanne").plot.bar(figsize=(20,5), x = "Nachname", y = "Lebensspanne", color ="blue", title = "Lebensspannen der Autoren")
```

##### Dazugehöriges Histogramm

```python
ger_childrens_book_authors["Lebensspanne"].plot.hist(x = "Lebensspanne in Jahre", bins = 25, title = "Verteilung der Lebensspannen")
```

Die Lebensspannen sind annähernd Normalverteilt, die Daten sind intervallskaliert und kontinuerlich.
